# This is my DataBase Homework 1

Import csv and set clean_data to empty

```python
import csv
clean_data = []
```

Read data into 'sales', which should contains a 'list' of 'dict'

```python
sales = list(map(dict,csv.DictReader(open('sale.csv', 'r'))))
```

This is the 'sales' data

```python
sales
```

    1)  loop to check if Product ID is already stored in the list clean_data
    2)  adding the total revenue if Product ID is already in the list
    3)  appending customer ID to the list of customer IDs for the respective           Product Id if it is already in list
    4)  adding all the information as it is to the list if Product ID is not in         the list already

```python
for i in sales:
        PI = i['Product ID']
        if PI:
            flag = 0
            for j in range(len(clean_data)): 
                if i['Product ID'] == clean_data[j]['Product ID']:
                    clean_data[j]['Total Revenue'] += float(i['Item Cost'])
                    clean_data[j]['Customer ID'].append(i['Customer ID']) 
                    flag = 1
            if not flag: 
                info = {'Product ID':i['Product ID'],'Customer ID':[i['Customer ID']],'Total Revenue':float(i['Item Cost'])}
            else:
                info = {}
            if info and info not in clean_data:
                clean_data.append(info)
```

    5)  loop to count no. of unique customers for each Product ID and remove           customer IDs from the list

```python
for i in range(len(clean_data)): 
        clean_data[i]['Customer Count'] = len(set(clean_data[i]['Customer ID']))
        del clean_data[i]['Customer ID']
```

    6)  loop to write values to the csv file

```python
with open('output.csv',mode='w',newline='',encoding ='utf8') as csvfile:
        fieldnames = ['Product ID','Customer Count','Total Revenue']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for e in clean_data:
            writer.writerow(e)
```
